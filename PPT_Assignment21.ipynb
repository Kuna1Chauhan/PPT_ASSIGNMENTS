```python
##Answer No.1 :-

class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def convertToBST(root):
    if not root:
        return

    # Perform in-order traversal and store node values in a list
    def inorderTraversal(node, values):
        if node:
            inorderTraversal(node.left, values)
            values.append(node.val)
            inorderTraversal(node.right, values)

    values = []
    inorderTraversal(root, values)

    # Sort the list of values
    values.sort()

    # Update node values in the binary tree with the sorted values
    def updateNodeValues(node, index):
        if node:
            updateNodeValues(node.left, index)
            node.val = values[index[0]]
            index[0] += 1
            updateNodeValues(node.right, index)

    updateNodeValues(root, [0])


# Create the binary tree
root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

# Print the original binary tree
def printBinaryTree(root):
    if root:
        print(root.val, end=" ")
        printBinaryTree(root.left)
        printBinaryTree(root.right)

print("Original Binary Tree:")
printBinaryTree(root)
print()

# Convert the binary tree to a binary search tree
convertToBST(root)

# Print the converted binary search tree
print("Converted Binary Search Tree:")
printBinaryTree(root)
```

```python
##Answer No.2 :-

class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def findDistance(root, node1, node2):
    if not root:
        return 0

    # Find the Lowest Common Ancestor (LCA) of the two nodes
    def findLCA(root, node1, node2):
        if not root:
            return None

        if root.val > node1.val and root.val > node2.val:
            return findLCA(root.left, node1, node2)

        if root.val < node1.val and root.val < node2.val:
            return findLCA(root.right, node1, node2)

        return root

    lca = findLCA(root, node1, node2)

    # Calculate the distance from LCA to node1
    def findDistanceToNode(node, target):
        if node.val == target.val:
            return 0

        if node.val > target.val:
            return 1 + findDistanceToNode(node.left, target)

        return 1 + findDistanceToNode(node.right, target)

    distance1 = findDistanceToNode(lca, node1)
    distance2 = findDistanceToNode(lca, node2)

    return distance1 + distance2


# Create the Binary Search Tree
root = TreeNode(6)
root.left = TreeNode(2)
root.right = TreeNode(8)
root.left.left = TreeNode(0)
root.left.right = TreeNode(4)
root.right.left = TreeNode(7)
root.right.right = TreeNode(9)
root.left.right.left = TreeNode(3)
root.left.right.right = TreeNode(5)

# Define the two nodes
node1 = TreeNode(2)
node2 = TreeNode(8)

# Find the distance between the two nodes in the BST
distance = findDistance(root, node1, node2)

# Print the distance
print("Distance between node1 and node2:", distance)
```

```python
##Answer No.4 :-

class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        self.next = None  # Pointer to the next node at the same level


def connectNodesAtSameLevel(root):
    if not root:
        return root

    queue = [root]  # Initialize the queue with the root node

    while queue:
        level_size = len(queue)

        for i in range(level_size):
            node = queue.pop(0)

            # Connect the current node to the next node at the same level
            if i < level_size - 1:
                node.next = queue[0]

            # Enqueue the left and right child nodes
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

    return root


# Create the binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

# Connect nodes at the same level in the binary tree
root = connectNodesAtSameLevel(root)

# Print the connected nodes at the same level
def printConnectedNodes(root):
    if not root:
        return

    node = root
    while node:
        print(node.val, end=" → ")
        node = node.next

        if not node:
            print("-1")
        else:
            print(node.val)

print("Connected Nodes at the Same Level:")
printConnectedNodes(root)
```
