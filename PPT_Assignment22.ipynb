```python
##Answer No.1 :-
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def convertToDLL(root):
    if not root:
        return None

    def inorderTraversal(node):
        nonlocal prev, head

        if node:
            inorderTraversal(node.left)

            if not head:
                head = node

            if prev:
                prev.right = node
                node.left = prev

            prev = node

            inorderTraversal(node.right)

    prev = None
    head = None
    inorderTraversal(root)

    # Set the left pointer of the head to None
    head.left = None

    return head

root = TreeNode(4)
root.left = TreeNode(2)
root.right = TreeNode(5)
root.left.left = TreeNode(1)
root.left.right = TreeNode(3)

head = convertToDLL(root)

# Print the doubly linked list from left to right
current = head
while current:
    print(current.val, end=" ")
    current = current.right
```

```python
##Answer No.2 :-

class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def flipBinaryTree(root):
    if not root or (not root.left and not root.right):
        return root

    flipped_left = flipBinaryTree(root.left)
    flipped_right = flipBinaryTree(root.right)

    root.left = None
    root.right = None

    if flipped_left:
        root.left = flipped_left
        flipped_left.right = root

    if flipped_right:
        root.right = flipped_right
        flipped_right.left = root

    return root


# Create the binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

# Print the original binary tree
def printBinaryTree(root):
    if root:
        print(root.val, end=" ")
        printBinaryTree(root.left)
        printBinaryTree(root.right)

print("Original Binary Tree:")
printBinaryTree(root)
print()

# Flip the binary tree
flipped_root = flipBinaryTree(root)

# Print the flipped binary tree
print("Flipped Binary Tree:")
printBinaryTree(flipped_root)
```

```python
##Answer No.3 :-

class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def printRootToLeafPaths(root):
    if not root:
        return

    stack = [(root, str(root.val))]

    while stack:
        node, path = stack.pop()

        if not node.left and not node.right:
            print(path)
        
        if node.left:
            stack.append((node.left, path + "->" + str(node.left.val)))
        
        if node.right:
            stack.append((node.right, path + "->" + str(node.right.val)))


# Create the binary tree
root = TreeNode(6)
root.left = TreeNode(3)
root.right = TreeNode(5)
root.left.left = TreeNode(2)
root.left.right = TreeNode(5)
root.right.right = TreeNode(4)
root.left.right.left = TreeNode(7)
root.left.right.right = TreeNode(4)

# Print all root-to-leaf paths
printRootToLeafPaths(root)
```

```python
##Answer No.4 :-

def checkTraversal(preorder, inorder, postorder):
    if not preorder and not inorder and not postorder:
        return True

    if len(preorder) != len(inorder) or len(preorder) != len(postorder):
        return False

    if set(preorder) != set(inorder) or set(preorder) != set(postorder):
        return False

    if len(preorder) == 1 and preorder[0] == inorder[0] == postorder[0]:
        return True

    root = preorder[0]
    root_idx_inorder = inorder.index(root)
    left_inorder = inorder[:root_idx_inorder]
    right_inorder = inorder[root_idx_inorder + 1:]

    left_preorder = preorder[1:1 + len(left_inorder)]
    right_preorder = preorder[1 + len(left_inorder):]

    left_postorder = postorder[:len(left_inorder)]
    right_postorder = postorder[len(left_inorder):-1]

    return (checkTraversal(left_preorder, left_inorder, left_postorder) and
            checkTraversal(right_preorder, right_inorder, right_postorder))


# Example usage
inorder = [4, 2, 5, 1, 3]
preorder = [1, 2, 4, 5, 3]
postorder = [4, 5, 2, 3, 1]

if checkTraversal(preorder, inorder, postorder):
    print("Yes")
else:
    print("No")
```
